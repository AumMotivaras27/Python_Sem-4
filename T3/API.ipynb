```python
import pandas as pd
import requests,json
api_key="fb197cd5b99d7ede47cd02216a9f9dfe"
city = input("Enter City Name:- ")
url = "http://api.openweathermap.org/geo/1.0/direct?q="+city+"&appid="+api_key
response = requests.get(url)
coordinates = response.json()
# print(coordinates)
lat = str(coordinates[0]['lat'])
lon = str(coordinates[0]['lon'])
print("Latitude = ",lat)
print("Longitude = ",lon)
print()
cw_url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}"
response = requests.get(cw_url)
current_wheather = response.json()
# print(json.dumps(current_wheather,indent=5))

wheather_description = current_wheather["weather"][0]["description"]
temperature = current_wheather["main"]["temp"]
pressure = current_wheather["main"]["pressure"]
humidity = current_wheather["main"]["humidity"]

print("\nWeather Details:")
print(f"Description: {wheather_description}")
print(f"Temperature: {temperature} °K")
print(f"Pressure: {pressure} hPa")
print(f"Humidity: {humidity} %")

ap_url = f"http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={api_key}"
response = requests.get(ap_url)
air_pollution = response.json()
# print(json.dumps(air_pollution,indent=5))

wf_url = f"http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}"
response = requests.get(wf_url)
weather_forcast= response.json()
# print(weather_forcast)
print()
D = {"Date_time":[],"Temp":[],"Pressure":[],"Humidity":[],"Weather":[]}
for i in weather_forcast['list']:
    D['Date_time'].append(i['dt_txt'])
    D['Temp'].append(i['main']['temp'])
    D['Pressure'].append(i['main']['pressure'])
    D['Humidity'].append(i['main']['humidity'])
    D['Weather'].append(i['weather'][0]['description'])
df = pd.DataFrame(D)
print(df)
print(df.head(40))
```
