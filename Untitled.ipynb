```python
# SUPERVISED LEARNING PROCESS 
import pandas as pd# defne the  problem
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
df=pd.read_csv('Book1.csv')# collect and prepre data
print(df)
x=df.iloc[:,0:1]# frame
print(type(x))
y=df.iloc[:,-1]# series
print(type(y))
plt.scatter(df['cgpa'],df['package'])
plt.show()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
m=lr.coef_
print(m)
c=lr.intercept_
print(c)
print(pd.DataFrame({'Actual pkg':y_test,'pred package':y_pred}))
print('mse=',mean_squared_error(y_test,y_pred))
print('rmse=',(mean_squared_error(y_test,y_pred))**0.5)
print('r2_score=',r2_score(y_test,y_pred))

```

```python
# newspaper ,radio and t.v. 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

df=pd.read_csv('advertising.csv')
x=df[['TV','Radio','Newspaper']]
y=df['Sales']
max1=0.0
for i in range(1,51):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    if r2_score(y_test,y_pred)>max1:
        max1=r2_score(y_test,y_pred)
        print(i)
print('mse=',mean_squared_error(y_test,y_pred))
print('r2_score=',max1)
print('rmse=',(mean_squared_error(y_test,y_pred))**0.5)

```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,export_text,plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
df=pd.read_csv('DecisionTreeDataset -Num.csv')
print(df)
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=4)
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=4)
dtree.fit(x_train,y_train)
print(dtree.get_depth())
print(export_text(dtree))
plt.figure(figsize=(25,20))
plot_tree(dtree,filled=True,feature_names=['CGPA','Communication','Apptitude','Programming Skill','Job Offered'],class_names=['No','Yes'])
plt.show()
y_pred=dtree.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,export_text,plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score
df=pd.read_csv('diabetes.csv')
print(df)
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=20)
dtree.fit(x_train,y_train)
print(dtree.get_depth())
print(export_text(dtree))
plt.figure(figsize=(25,20))
plot_tree(dtree,filled=True,feature_names=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age' ],class_names=['0','1'])
plt.show()
y_pred=dtree.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
