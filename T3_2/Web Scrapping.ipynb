```python
# movie title,rating

from bs4 import BeautifulSoup
import pandas as pd

f=open("30 Most Popular Movies Right Now_ What to Watch In Theaters and Streaming __ Rotten Tomatoes – Movie and TV News.html","r",encoding="utf-8")
response = f.read()
soup = BeautifulSoup(response,"html.parser")
# print(soup.prettify())
movie = soup.find_all("div",class_="article_movie_title")
titles = []
ratings = []
for i in movie:
    t = i.find("h2")
    title = i.find("a").text.strip()
    rating = i.find("span",class_="tMeterScore").text.strip()
    titles.append(title)
    ratings.append(rating)
    
data = {"Movie Title":titles,"Rating":ratings}
df = pd.DataFrame(data)
df
```

```python
# name,rating,fprice,mprice

from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

f=open("flipkart.html","r",encoding="utf-8")
response = f.read()
soup = BeautifulSoup(response,"html.parser")
# print(soup.prettify())
flip = soup.find_all("div",class_="_3pLy-c")
name = []
rating = []
fprice = []
mprice = []
for i in flip:
    t = i.find("div")
    n = i.find("div",class_="_4rR01T").text.strip()
    name.append(n)
    r = i.find("div",class_="_3LWZlK").text.strip()
    rating.append(r)
    fp = i.find("div",class_="_30jeq3 _1_WHN1").text.strip()
    fprice.append(fp)
    try:
        mp = i.find("div",class_="_3I9_wc _27UcVY").text.strip()
        mprice.append(mp)
    except:
        mprice.append(np.nan)
    
data = {"Mobile Name":name,"Rating":rating,"Flipkart Price":fprice,"Market Price":mprice}
df = pd.DataFrame(data)
df
```

```python
# Extract values of all cars and create data frame for Car name,mileage,dealer name,rating,review counts,car price for
# Certified used Mercedes-Benz for sale.html

from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

f=open("Certified used Mercedes-Benz for sale.html","r",encoding="utf-8")
response = f.read()
soup = BeautifulSoup(response,"html.parser")
# print(soup.prettify())
car_data = soup.find_all("div",class_="vehicle-details")
cname = []
mil = []
dname = []
rat = []
rev_cnt = []
cprice = []
for i in car_data:
    t = i.find("div")
    cn = i.find("h2",class_="title").text.strip()
    cname.append(cn)
    m = i.find("div",class_="mileage").text.strip()
    mil.append(m)
    dn = i.find("div",class_="dealer-name").text.strip()
    dname.append(dn)
    try:
        r = i.find("span",class_="sds-rating__count").text.strip()
        rat.append(r)
    except:
        rat.append(np.nan)
    rc = i.find("span",class_="sds-rating__link").text[1:-1].strip()
    rev_cnt.append(rc)
    cp = i.find("span",class_="primary-price").text.strip()
    cprice.append(cp)

data = {"Car Name":cname,"Mileage":mil,"Dealer Name":dname,"Rating":rat,"Review Count":rev_cnt,"Car Price":cprice}
df = pd.DataFrame(data)
df
```
