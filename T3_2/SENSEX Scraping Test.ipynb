```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

with open(
    "SENSEX Top Gainers, BSE Daily Top Gainers - Stock Market.html",
    "r",
    encoding="utf-8",
) as file:
    response = file.read()
    soup = BeautifulSoup(response, "html.parser")
    # print(soup.prettify())
    articles = soup.find_all("tr")
    frame = []
    for a in articles:
        try:
            Company = a.find("td", class_="v-align").text.strip()
            Current_Price = a.find("td", class_="txt-right v-align").text.strip()
            Prev_Close_Date = a.find("td", class_="txt-right no-wrap").text.strip()
            Change = a.find("td", class_="txt-ctr").text.strip()
            Change_Percent = a.find("span", class_="green-btn").text.strip()
            day_low = a.find("span", class_="fl f-9").text.strip()
            day_high = a.find("span", class_="fr f-9").text.strip()
        except AttributeError:
            continue

        frame.append(
            (
                Company,
                Current_Price,
                Prev_Close_Date,
                Change,
                Change_Percent,
                day_low,
                day_high,
            )
        )

df = pd.DataFrame(
    frame,
    columns=[
        "Company",
        "Current Price",
        "Prev Close Date",
        "Change",
        "Change Percent(%)",
        "Day Low",
        "Day High",
    ],
)
df
# df.to_csv("SENSEX_Top_Gainers.csv", index=False)
```
